```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets

def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'],columns = ds['feature_names'])
    code_species_map = dict(zip(range(3), ds['target_names']))
    df['species'] =[code_species_map[c] for c in ds['target']]
    return df
df = get_iris_df()
df
```

```python
#chart 1
sums_by_species = df.groupby('species').sum()

adf = sums_by_species
print(adf)
```

```python
#chart 2
sums_by_species = df.groupby('species').sum()
#var = 'sepal width (cm)'
sums_by_species.plot(kind='pie',fontsize =4, subplots=True)
plt.title('Total Measurements,by species')
plt.savefig('iris_pie_for_each_variable.jpg')
plt.show()
plt.close()
```

```python
#bar chart
sums_by_species = df.groupby('species').sum()
var = 'sepal width (cm)'
sums_by_species[var].plot(kind='bar',fontsize =15, rot=30)
plt.title('Breaking for' + var,fontsize=20)
plt.savefig('iris_bar_for_one_variable.jpg')
plt.show()
plt.close()
```

```python
#bar chart
sums_by_species = df.groupby('species').sum()
#var = 'sepal width (cm)'
sums_by_species.plot(kind='bar',subplots = True, fontsize = 10 )
plt.suptitle('Total measurements by species',fontsize=13)
plt.show()
plt.close()
```

```python
df.plot(kind='hist',subplots=True, layout=(2,2))
plt.suptitle('iris Histograms', fontsize=20)
plt.show()
```

```python
for spec in df['species'].unique():
    forspec = df[df['species']==spec]
    forspec['petal length (cm)'].plot(kind='hist',alpha=0.4,label=spec)
plt.legend(loc='upper right')
plt.suptitle('Petal Length by Species')
plt.savefig('iris_hist_by_spec.jpg')
plt.show()
```

```python
#Means sd
col = df['petal length (cm)']
Average = col.mean()
print(Average)

std = col.std()
print(std)

median = col.quantile(0.5)
print(median)

percentile25 = col.quantile(0.25)
print(percentile25)
```

```python
col = df['petal length (cm)']
per25 = col.quantile(0.25)
per75 = col.quantile(0.75)
Clean_avg = col[(col>per25)&(col<per75)].mean()
print(per25, per75, Clean_avg)
```

```python
#box plot
col = "sepal length (cm)"
df['ind'] = pd.Series(df.index).apply(lambda i:i% 50)
df.pivot('ind','species')[col].plot(kind='box')
plt.show()
plt.close()
```

```python
#scatter plot

df.plot(kind="scatter",
        x = "sepal length (cm)",y="sepal width (cm)", s= 20)
plt.title("Length vs Width")
plt.show()
plt.close()
```

```python
colors = ["r","g","b"]
markers = [".","*","^"]
fig,ax = plt.subplots(1,1)
for i, spec in enumerate(df['species'].unique()):
    ddf = df[df['species']==spec]
    ddf.plot(kind="scatter",
    x="sepal width (cm)", y="sepal length (cm)",
            alpha=0.5, s =10*(i+1), ax=ax,
            color=colors[i],marker=markers[i],label=spec)
plt.legend()
plt.show()
```
